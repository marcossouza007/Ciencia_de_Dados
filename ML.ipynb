```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
X = df.drop('produtividade', axis=1)
y = df['produtividade']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

print('RMSE LR:', np.sqrt(mean_squared_error(y_test, pred_lr)))
print('R2 LR:', r2_score(y_test, pred_lr))
```

```python
# Random Forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

print('RMSE RF:', np.sqrt(mean_squared_error(y_test, pred_rf)))
print('R2 RF:', r2_score(y_test, pred_rf))
```
